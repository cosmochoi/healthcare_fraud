```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
outpatient_test = pd.read_csv("./dataset/Test_Outpatientdata-1542969243754.csv")
```

```python
outpatient_test.head()
```

```python
outpatient_train = pd.read_csv("./dataset/Train_Outpatientdata-1542865627584.csv")
```

```python
outpatient_train.head()
```

```python
outpatient = pd.concat((outpatient_train, outpatient_test)).reset_index(drop=True)
```

```python
outpatient.head()
```

```python
#Doctors
outpatient['AttendingPhysician'].nunique()
```

```python
s = outpatient['AttendingPhysician'].value_counts().nlargest(20)
s.plot.bar()
plt.title('Outpatient Medical Doctors')
```

```python
outpatient['BeneID'].nunique()
```

```python
b = outpatient['BeneID'].value_counts().nlargest(20)
b.plot.bar()
plt.title('Outpatients (BeneID)')
```

```python
#Each physician works for # many hospitals
hosPhy = pd.DataFrame(outpatient.groupby('AttendingPhysician')['Provider'].nunique()).reset_index()
hosPhy.columns = ["Physician", "Hospitals"]
percent = hosPhy[hosPhy.Hospitals > 1].shape[0]/hosPhy.shape[0] # Percentage of doctors who work for 2 or more hospitals
print("Percent Physicians associated with two or more hospitals:\n" + '%.3f' %percent)
hosPhy[hosPhy.Hospitals > 1]
```

```python
#Each patient goes to # many hospitals
hosPat = pd.DataFrame(outpatient.groupby('BeneID')['Provider'].nunique()).reset_index()
hosPat.columns = ['Patient', 'Hospital']
patperc = hosPat[hosPat.Hospital > 1].shape[0]/hosPat.shape[0] # Percentage of doctors who work for 2 or more hospitals
print("Percent Patients that go to two or more hospitals:\n" + '%.3f' %patperc)
hosPat[hosPat.Hospital > 1]
```

```python
from datetime import datetime
outpatient['ClaimStartDt'] = [datetime.strptime(date, '%Y-%m-%d') for date in outpatient.ClaimStartDt]
```

```python
outpatient['ClaimEndDt'] = [datetime.strptime(date, '%Y-%m-%d') for date in outpatient.ClaimEndDt]
```

```python
outpatient['duration'] = outpatient['ClaimEndDt'] - outpatient['ClaimStartDt']
outpatient['duration'] = list(map(lambda x: x.days, outpatient['duration']))
```

```python
outpatient['month'] = list(map(lambda x: x.month, outpatient['ClaimStartDt']))
```
